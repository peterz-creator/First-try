```python
# 安装获取财经数据的库 (Install the library)
!pip install yfinance

import yfinance as yf

# 设置你想查看的资产代码，比如 NVDA (英伟达) 或 BTC-USD (比特币)
ticker_symbol = "NVDA"

# 获取数据
data = yf.Ticker(ticker_symbol)
current_price = data.history(period="1d")['Close'].iloc[-1]

print(f"{ticker_symbol} 的当前价格是: ${current_price:.2f}")
```
